```python
import pandas as pd
df=pd.read_csv("salaries.csv")
df.head()
```

```python
inputs=df.drop('salary_more_then_100k',axis='columns')
target=df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder
le_job=LabelEncoder
le_degree=LabelEncoder
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'],df['salary_more_then_100k'])
inputs['job_n'] = le_company.fit_transform(inputs['job'],df['salary_more_then_100k'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'],df['salary_more_then_100k'])
inputs
```

```python
inputs_n=inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```
